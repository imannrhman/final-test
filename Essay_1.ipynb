```python
import pandas as pd
import requests
from datetime import datetime
```

```python
base_url = "https://berita-indo-api-next.vercel.app/api/"

links_berita = [
    "cnn-news/teknologi",
    "antara-news/terkini",
    "cnbc-news/entrepreneur",
    "cnbc-news/entrepreneur",
    "republika-news/sepakbola",
    "tempo-news/nasional",
    "okezone-news/lifestyle",
    "kumparan-news",
    "tribun-news/jabar/"
]
```

```python
def cleanCreator(value) :
  result = value.replace("-", " ")
  result = result.replace("/", " ")


  return result
```

```python
df_news = []
for news in links_berita :
  response = requests.get(base_url + news)
  results = response.json()
  new_dataframe = pd.DataFrame(results['data'])
  new_dataframe['creator'] = cleanCreator(news)

  df_news.append(new_dataframe)


dataframe = pd.concat(df_news)
```

```python
dataframe = dataframe[["title", "link", "contentSnippet", "contentSnippet", "image", "creator"]]
dataframe.head()

```

```python
dataframe['isoDate'] = dataframe.isoDate.apply(lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ"))
```
